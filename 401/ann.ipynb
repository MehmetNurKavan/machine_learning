Veri seti hikayesi ve problem: Şeker hastalığı tespiti

```python
df = pd.read_csv("./diabetes.csv")
df.head()
```

ANN

```python
y= df["Outcome"]
x= df.drop(["Outcome"], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train= scaler.transform(x_train)
```

```python
scaler.fit(x_test)
x_test = scaler.transform(x_test)
```

model tuning

```python
mlpc_parms = {"alpha":[1,5,0.1,0.01,0.03,0.005,0.0001],"hidden_layer_sizes": [(10,10),(100,100,100),(100,100),(3,5)]}
```

```python
mlpc= MLPClassifier(solver = "lbfgs", activation="logistic")
```

```python
mlpc_cv_model = GridSearchCV(mlpc, mlpc_parms, cv=10, n_jobs=-1,verbose=2).fit(x_train, y_train)
```

```python
mlpc_cv_model
```

```python
mlpc_cv_model.best_params_
```

final model

```python
mlpc_tuned = MLPClassifier(
    solver="lbfgs",
    activation='logistic',
    alpha=5,
    hidden_layer_sizes=(100, 100)
).fit(x_train, y_train)
```

```python
y_pred=mlpc_tuned.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
