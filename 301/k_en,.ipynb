```python
df=pd.read_csv('C:/Users/Mehmet Nur Kavan/Desktop/machine_learning/301/Hitters.csv')
df=df.dropna()
dms=pd.get_dummies(df[['League','Division','NewLeague']])
y=df["Salary"] # bağımlı değişken
x_=df.drop(['Salary','League','Division','NewLeague'], axis=1).astype('float64')
x=pd.concat([x_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

Model Tuning

```python
knn_model
```

```python
range(10)
```

```python
RMSE=[]

for k in range(10):
    k=k+1
    knn_model = KNeighborsRegressor(n_neighbors = k).fit(x_train, y_train)
    #dongu yaparak birbirinden farklı algoritmalar yapıyrouz
    y_pred= knn_model.predict(x_test)
    rmse= np.sqrt(mean_squared_error(y_test, y_pred))
    # gercek degerlerle tahmın edilen degerlerin farkını aldık
    RMSE.append(rmse)
    print("k=", k, " için RMSE degeri:", rmse)
```

GridSearchCV

belirlemeye calsıtıgımız hiperparametre fonksiyon degerkelriimiz kıyaslamka icın kullanıyoruz

```python
knn_params= {"n_neighbors": np.arange(1, 30, 1)}
```

```python
knn= KNeighborsRegressor()
```

```python
knn_cv_model= GridSearchCV(knn, knn_params, cv=10).fit(x_train, y_train)
```

```python
knn_cv_model.best_params_
```

8 en iyisiymiş

final model

```python
knn_tuned = KNeighborsRegressor(n_neighbors=knn_cv_model.best_params_["n_neighbors"]).fit(x_train, y_train)
```

```python
y_pred = knn_tuned.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```
